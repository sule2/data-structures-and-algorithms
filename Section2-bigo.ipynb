```python
def bigon(n):
    for i in range(0,n):
        print(i)
```

```python
bigon(5)
```

```python
def bigon2(n):
    for i in range(0,n):
        for j in range(0,n):
            print(i,j)
```

```python
bigon2(5)
```

```python
def bigon3(n):
    for i in range(0,n):
        for j in range(0,n):
            for k in range(0,n):
                print(i,j,k) #n3
```

```python
import math
def logn(n):
    while n>1:
        n=math.floor(n/2)
        print(n)
```

```python
logn(8)
```

```python
 def nlogn(n):
        lim = n
        while n>1:
            n=math.floor(n/2)
            for i in range(1,lim):
                print(i)
```

```python
nlogn(8)
```

```python
def nfactorial(n): #O(n!)
     if n == 0:
        print("1")
        return
     else:
        for i in range(0,n):
            print(i)
            nfactorial(n-1) #recursive
```

```python
nfactorial(5)
```

```python
counter=0
def actualfactorial(n): #O(n)
    global counter
    counter+=1
    print(counter)
    if n == 1 or n == 0 :
        return 1
    else:
        return n*actualfactorial(n-1)
```

```python
actualfactorial(5)
```
